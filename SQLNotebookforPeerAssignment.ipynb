##### Display the names of the unique launch sites  in the space mission

```python
query='SELECT DISTINCT Launch_Site from SPACEXTBL'
cur.execute(query)
launch_sites = cur.fetchall()

for site in launch_sites:
    print(site[0])
```


#####  Displaying 5 records where launch sites begin with the string 'CCA' 




```python
query= "SELECT * from SPACEXTBL where Launch_Site like 'CCA%' limit 5"
cur.execute(query)
launch_sites = cur.fetchall()

for site in launch_sites:
    print(site[0])
```

##### the total payload mass carried by boosters launched by NASA (CRS)


```python
query="Select sum(PAYLOAD_MASS__KG_) from SPACEXTBL where Customer='NASA (CRS)' "
cur.execute(query)
total_payloads=cur.fetchall()
total_payloads
```






##### Displaying average payload mass carried by booster version F9 v1.1


```python
query="Select avg(PAYLOAD_MASS__KG_) from SPACEXTBL where Booster_Version='F9 v1.1' "
cur.execute(query)
avg_payload=cur.fetchall()
avg_payload
```



##### Listing the date when the first succesful landing outcome in ground pad was acheived.
 


```python
query="Select min(date) from SPACEXTBL where Landing_Outcome like 'Success (ground pad)' "
cur.execute(query)
min_succes_landing=cur.fetchall()
min_succes_landing
```



##### Listing the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
query="Select Customer from SPACEXTBL where Landing_Outcome = 'Success (drone ship)' and PAYLOAD_MASS__KG_>4000 and PAYLOAD_MASS__KG_<6000 "
cur.execute(query)
min_succes_landing=cur.fetchall()
min_succes_landing
```



##### Listing the total number of successful and failure mission outcomes


```python
query = "SELECT mission_outcome, COUNT(*) AS total_count FROM SPACEXTBL GROUP BY mission_outcome"
cur.execute(query)
min_succes_landing=cur.fetchall()
min_succes_landing
```



##### Listing the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
query = """SELECT Booster_Version FROM SPACEXTBL where PAYLOAD_MASS__KG_ =(
    select max(PAYLOAD_MASS__KG_) from SPACEXTBL)"""
cur.execute(query)
Booster_Version=cur.fetchall()
Booster_Version
```
